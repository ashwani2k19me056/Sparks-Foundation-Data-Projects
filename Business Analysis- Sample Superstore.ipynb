## Profit vs Discount

```python
sns.scatterplot(x = 'Profit', y='Discount', data = df)
```

```python
sns.lineplot(y = 'Profit', x='Discount', data = df)
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x='Sales', y='Profit', data = df, hue = 'Discount')
```

### Observation 4

There is no direct relation between discount and profits. Initially as the discount increases upto 20 percent the profits are positive and it increases. But after the discount of more than 20 percent, the profits decline. 

### Segment Profit/Sales



```python
plt.figure(figsize=(16,8))
df.groupby('Segment')[['Profit', 'Sales']].sum().plot.bar(color=['Red','Orange'], figsize=(12,6))
plt.ylabel('Profit / Sales')
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Quantity', y='Profit', data = df)
```

### Observation 5 

The profits and sales are maximum in Consumer segment and minimum in Home Office segment.

### Segement wise sales analysis

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Region', y='Sales', data = df, hue ='Segment', palette='rocket' )
plt.title('Segment wise sales in each region')
```

### Observation 6 

The segment wise sales in each region is almost similar

### Sales in each region

```python
plt.figure(figsize=(12,6))
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(figsize=(12,6))
```

### Observation 6

The profits and sales are maximum in West region and minimum in South region

### Profit/Loss Sales across States

```python
df1 = df.groupby('State')[['Profit','Sales']].sum().sort_values(by='Sales', ascending= False)
df1.plot.bar(color=['Blue', 'Orange'], figsize=(14,10))
plt.ylabel('Profit / Sales')
plt.title('Profit/Loss and Sales across States')
```

### Observation 8

The Sales and Profits are maximum in the states of California and New York.
The States of Texas, Pennsylvania and Ohio on the other hand results in overall loss. 

### Category Analysis

```python
df.groupby('Category')[['Profit','Sales']].sum().sort_values(by='Sales', ascending= False).plot.bar(color=['Blue','Orange'], figsize=(14,8))
plt.ylabel('Profits / Sales')
```

### Observation 9

Technology and Office Supplies have high profit.
Furniture have less profit

### Sub-Category Analysis

```python
df.groupby('Sub-Category')[['Profit','Sales']].sum().sort_values(by='Sales', ascending= False).plot.bar(color=['Blue','Orange'], figsize=(12,6))
```

```python
plt.figure(figsize=(16,8))
sns.countplot(df['Sub-Category'])
```

```python
plt.figure(figsize=(16,8))
sns.boxplot(x='Sub-Category', y='Profit', data =  df)
```

### Observation 10

● Phones and Chairs have highest sales.

● Binders and Papers have highest numbers of sales with correspondingly high profits. Binders also have large losses.

● Tables and Bookcases are resulting in overall losses.
